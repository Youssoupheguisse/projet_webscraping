# fonction de séparation des pays est des valeurs

```python
# fonction de séparation 
def separate_dataframe(dataframe, columns = []):
  liste = dataframe.to_list()
  dicte = {column: [] for column in columns} 
  j = 0
  for element in enumerate(liste):
    dicte[columns[j]] = dicte[columns[j]] + [element[1]]
    if j == len(columns)-1:
      j = 0
    else:
      j += 1
  return dicte
```

```python
list1df.iloc[:, 0].to_list()
list1df = pd.DataFrame(separate_dataframe(list1df.iloc[:,0], columns = ["Pays", "Population"]))
list1df
```
